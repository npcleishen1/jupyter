```python
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
import openpyxl
import akshare as ak
InteractiveShell.ast_node_interactivity = "all"
pd.options.display.max_columns = 1000
pd.options.display.max_rows = 5000
pd.options.display.float_format = lambda x: '%.5f' % x
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
%config InlineBackend.figure_format='svg'
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from datetime import datetime
```

```python
#date1=''#计算起始日期
date2='20241231'#最终日期
date3='20130820'#策略起始日期
date3=datetime.strptime(date3,"%Y%m%d")
date2=datetime.strptime(date2,"%Y%m%d")
m_days=250
n_days=(date2-date3).days
CSI300=ak.stock_zh_index_daily(symbol='sh000300')
CSI300['date']=pd.to_datetime(CSI300['date'])
CSI300=CSI300[(CSI300['date']<=date2)&(CSI300['date']>=date3)]
CSI300=CSI300.dropna().reset_index(drop=True)
total_return1=(CSI300.loc[len(CSI300)-1,'close']-CSI300.loc[0,'close'])/CSI300.loc[0,'close']
annual_return1=(total_return1+1)**(m_days/n_days)-1
CSI300['max']=CSI300['high'].cummax()
CSI300['back']=CSI300['max']-CSI300['close']
CSI300['return']=CSI300['close']-CSI300.loc[0,'close']
CSI300['day_return'] = CSI300['close'] / CSI300['close'].shift(1) - 1
CSI300.loc[0, 'day_return'] = 0 
back1=CSI300['back'].max()/CSI300.loc[len(CSI300)-1,'max']
volatility1=CSI300['day_return'].std()*np.sqrt(m_days)
sharpe1=(annual_return1-0.02)/volatility1
df=pd.DataFrame(columns=['总收益率','最大回撤','回撤比','年化收益率','波动率','夏普比率'],index=['沪深300','恒生指数','标普500'])
df.loc['沪深300','总收益率']=total_return1
df.loc['沪深300','最大回撤']=-back1
df.loc['沪深300','回撤比']=back1/total_return1
df.loc['沪深300','年化收益率']=annual_return1
df.loc['沪深300','波动率']=volatility1
df.loc['沪深300','夏普比率']=sharpe1
```
